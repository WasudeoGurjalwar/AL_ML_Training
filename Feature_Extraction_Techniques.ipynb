Feature extraction techniques in machine learning are methods used to transform the original set of features (variables) into a new set of features with reduced dimensionality while retaining as much relevant information as possible. These techniques are primarily used for dimensionality reduction, noise reduction, and creating a more manageable representation of data. Here's a brief technical overview of feature extraction techniques:

1. **Principal Component Analysis (PCA)**:
   - **Objective**: To find linear combinations of the original features (principal components) that capture the maximum variance in the data.
   - **Technique**: PCA identifies orthogonal axes (principal components) along which the data varies the most. The first principal component captures the most variance, the second captures the second most, and so on.
   - **Application**: Reducing dimensionality in high-dimensional datasets, decorrelating features, and eliminating multicollinearity.

2. **Linear Discriminant Analysis (LDA)**:
   - **Objective**: To find linear combinations of features that maximize the separation between different classes in a supervised setting.
   - **Technique**: LDA identifies directions (linear discriminants) that maximize the ratio of between-class variance to within-class variance.
   - **Application**: Dimensionality reduction for classification tasks, finding discriminative features.

3. **Non-Negative Matrix Factorization (NMF)**:
   - **Objective**: To factorize a non-negative data matrix into two non-negative matrices with reduced dimensionality.
   - **Technique**: NMF aims to express data as a linear combination of basis vectors with non-negative coefficients.
   - **Application**: Topic modeling, image processing, and text mining.

4. **Autoencoders**:
   - **Objective**: To learn a compressed representation of the input data through neural networks.
   - **Technique**: Autoencoders consist of an encoder that maps input data to a lower-dimensional representation and a decoder that reconstructs the input from the lower-dimensional representation.
   - **Application**: Image and text data compression, anomaly detection, and feature learning.

5. **t-Distributed Stochastic Neighbor Embedding (t-SNE)**:
   - **Objective**: To visualize high-dimensional data in lower dimensions while preserving the pairwise similarity relationships.
   - **Technique**: t-SNE minimizes the divergence between probability distributions over pairs of data points in high-dimensional and low-dimensional spaces.
   - **Application**: Data visualization and clustering assessment.

6. **Factor Analysis**:
   - **Objective**: To explain the variance-covariance structure of observed variables by a smaller number of latent factors.
   - **Technique**: Factor analysis models the relationships between observed variables and latent factors using a factor loading matrix.
   - **Application**: Psychology research, finance, and social sciences.

7. **Independent Component Analysis (ICA)**:
   - **Objective**: To separate a multivariate signal into additive, independent components.
   - **Technique**: ICA finds linear combinations of observed variables to maximize their statistical independence.
   - **Application**: Blind source separation, image processing, and signal processing.

These feature extraction techniques are valuable tools for preprocessing data, reducing dimensionality, and creating more informative representations of complex datasets. The choice of technique depends on the specific goals of the machine learning task and the nature of the data.

## Example 1  - Principal Component Analysis (PCA)

Demonstrating Principal Component Analysis (PCA) using the Breast Cancer dataset from scikit-learn:

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the Breast Cancer dataset as an example
data = load_breast_cancer()
X = data.data
y = data.target
print("---------------")
# Get the number of rows (samples) and columns in the dataset
num_rows = data.data.shape[0]
num_colns = data.data.shape[1]
print(num_rows, num_colns)
print("---------------")

# Standardize the features (mean=0, variance=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Instantiate PCA with the desired number of components (e.g., 2)
pca = PCA(n_components=2)

# Fit and transform the data using PCA
X_pca = pca.fit_transform(X_scaled)

# Create a DataFrame with the principal components
df_pca = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
df_pca['target'] = y

# Display the first few rows of the principal components
print("Principal Components:")
print(df_pca.head())

# Visualize the data in the new feature space
plt.figure(figsize=(8, 6))
plt.scatter(df_pca['PC1'][df_pca['target'] == 0], df_pca['PC2'][df_pca['target'] == 0], label='Class 0', alpha=0.5)
plt.scatter(df_pca['PC1'][df_pca['target'] == 1], df_pca['PC2'][df_pca['target'] == 1], label='Class 1', alpha=0.5)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Breast Cancer Dataset')
plt.legend()
plt.grid()
plt.show()
```

## Example 2 - Linear Discriminant Analysis (LDA)

Instead of PCA, we can use Linear Discriminant Analysis (LDA) for dimensionality reduction and feature extraction. LDA is particularly useful when you have a classification problem and want to find a lower-dimensional representation that maximizes class separability. Here's the code using LDA:

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt

# Load the Breast Cancer dataset as an example
data = load_breast_cancer()
X = data.data
y = data.target

# Standardize the features (mean=0, variance=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Instantiate LDA with the desired number of components (e.g., 1)
lda = LinearDiscriminantAnalysis(n_components=1)

# Fit and transform the data using LDA
X_lda = lda.fit_transform(X_scaled, y)

# Create a DataFrame with the LDA components
df_lda = pd.DataFrame(data=X_lda, columns=['LDA1'])
df_lda['target'] = y

# Display the first few rows of the LDA components
print("LDA Components:")
print(df_lda.head())

# Visualize the data in the new feature space
plt.figure(figsize=(8, 6))
plt.scatter(df_lda['LDA1'][df_lda['target'] == 0], np.zeros_like(df_lda['LDA1'][df_lda['target'] == 0]), label='Class 0', alpha=0.7)
plt.scatter(df_lda['LDA1'][df_lda['target'] == 1], np.zeros_like(df_lda['LDA1'][df_lda['target'] == 1]), label='Class 1', alpha=0.7)
plt.xlabel('LDA Component 1')
plt.title('LDA of Breast Cancer Dataset')
plt.legend()
plt.grid()
plt.show()
```

The error you encounter with `lda = LinearDiscriminantAnalysis(n_components=2)` occurrs because the Breast Cancer dataset, as provided by scikit-learn's `load_breast_cancer` function, has two classes (`n_classes = 2`) and 30 features (`n_features = 30`).

For LDA, the number of components (`n_components`) you can specify is constrained by the number of classes and features:

- In general, `n_components` cannot be greater than `min(n_features, n_classes - 1)`. This constraint arises because the maximum number of discriminative components that can be extracted is limited by the minimum of the number of features and the number of classes minus one.

In the case of the Breast Cancer dataset:

- `n_classes` is 2 because there are two classes (0 and 1).
- `n_features` is 30 because there are 30 feature columns in the dataset.

Therefore, the maximum allowable value for `n_components` in this specific dataset is `min(30, 2 - 1)`, which is 1. This means you can only extract one linear discriminant component for this dataset.

If you attempt to set `n_components` to 2 in this dataset, you will receive the "ValueError: n_components cannot be larger than min(n_features, n_classes - 1)" error because you are trying to extract more components than allowed by the data's characteristics.

To summarize, for the Breast Cancer dataset and similar binary classification datasets, LDA can be applied with `n_components` set to 1, which allows you to extract a single linear discriminant component. If you wish to visualize the data in a two-dimensional space, PCA or other techniques may be more appropriate.

## Example 3 - Non-Negative Matrix Factorization (NMF)

Non-Negative Matrix Factorization (NMF) is often used for topic modeling, which is a technique to discover topics within a collection of documents. Here's a coded example of topic modeling using NMF with Python's scikit-learn library:

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF

# Sample documents (replace with your own corpus)
machine_learning_documents = [
    "Machine learning is a subset of artificial intelligence.",
    "Deep learning is a popular branch of machine learning.",
    "Supervised and unsupervised learning are common ML techniques.",
    "NLP and computer vision are applications of ML.",
    "ML models can make predictions based on data."
]

football_documents = [
    "Football is a widely popular sport played globally.",
    "Soccer, known as football in most countries, is loved by fans.",
    "FIFA World Cup is a major football tournament.",
    "Football matches are held in stadiums with cheering fans.",
    "Many players aspire to become professional footballers."
]

# Combine the documents into a single corpus
documents = machine_learning_documents + football_documents

# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')

# Fit and transform the documents into TF-IDF vectors
tfidf_matrix = vectorizer.fit_transform(documents)

# Define the number of topics (components) for NMF
n_topics = 2

# Instantiate NMF with the desired number of topics
nmf = NMF(n_components=n_topics, random_state=42)

# Fit NMF to the TF-IDF matrix
nmf.fit(tfidf_matrix)

# Display the top words for each topic
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(nmf.components_):
    top_words_indices = topic.argsort()[:-10:-1]
    top_words = [feature_names[i] for i in top_words_indices]
    print(f"Topic {topic_idx + 1}: {', '.join(top_words)}")

# Transform the documents into topic distributions
topic_distributions = nmf.transform(tfidf_matrix)

# Create a DataFrame to show the topic distribution for each document
df_topic_distribution = pd.DataFrame(topic_distributions, columns=[f"Topic {i + 1}" for i in range(n_topics)])
df_topic_distribution.index = [f"Document {i + 1}" for i in range(len(documents))]
print("\nTopic Distribution for Each Document:")
print(df_topic_distribution)
```

#### **Explaination to some key Pointers**

In the above code:

> We have two sets of documents: machine_learning_documents and football_documents, each containing documents related to their respective topics.

> We combine these documents into a single corpus (documents) by concatenating the two lists.

> We perform TF-IDF vectorization, NMF for topic modeling, and displays the top words for each topic and the topic distribution for each document.

> The line of code `top_words_indices = topic.argsort()[:-10:-1]` is used to find the indices of the top words within a topic based on their importance scores. Let me break down how this line works:

1. `topic` is an array that represents the importance scores of words in a specific topic. Each element of the `topic` array corresponds to the importance score of a word in that topic.

2. `argsort()` is a NumPy function that returns the indices that would sort an array in ascending order. By default, it sorts in ascending order.

3. `topic.argsort()` returns an array of indices that would sort the `topic` array in ascending order of importance scores.

4. `[:-10:-1]` is slicing notation applied to the sorted indices array. Here's what each part does:
   - `[:-10]`: This part slices the array from the beginning up to (but not including) the 10th element from the end. In other words, it excludes the top 10 indices.
   - `[::-1]`: This part reverses the order of the sliced array. So, it effectively selects the top 10 indices with the highest importance scores in descending order.

As a result, `top_words_indices` contains the indices of the top 10 words within the topic, sorted by their importance scores from highest to lowest. These indices can be used to retrieve the actual words from the vocabulary or feature names to understand which words are most relevant to the topic.



### Quick Note of max_df and min_df

In the `TfidfVectorizer` of scikit-learn, the `max_df` and `min_df` parameters are used to control the inclusion of terms (words or tokens) in the TF-IDF matrix based on their document frequency within the corpus:

1. `max_df` (maximum document frequency): It specifies a threshold such that terms with a document frequency higher than this value will be ignored. In other words, terms that occur in more than a certain percentage of documents will be excluded from the TF-IDF matrix. For example, if `max_df=0.95`, terms that appear in more than 95% of the documents will be removed. This can help in removing very common terms that may not be informative.

2. `min_df` (minimum document frequency): It specifies a threshold for the minimum number of documents in which a term must appear to be included in the TF-IDF matrix. Terms that occur in fewer documents than this threshold will be ignored. For example, if `min_df=2`, terms must appear in at least 2 documents to be considered. This can help in removing very rare terms that may not be useful for analysis.

By setting appropriate values for `max_df` and `min_df`, you can control the vocabulary size and the terms included in the TF-IDF matrix, which can affect the quality of your text analysis. These parameters are useful for filtering out noisy or uninformative terms and tailoring the representation of text data to your specific analysis needs.

## Example 4 - t-Distributed Stochastic Neighbor Embedding (t-SNE)

t-Distributed Stochastic Neighbor Embedding (t-SNE) is a dimensionality reduction technique often used for visualizing high-dimensional data in a lower-dimensional space while preserving the pairwise similarity relationships between data points. It's commonly used to visualize the quality of clustering results. Here's a coded example demonstrating how t-SNE can be used for this purpose:

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Generate synthetic data with three well-separated clusters
X, y = make_blobs(n_samples=300, centers=3, random_state=42, cluster_std=1.0)
## A higher value of cluster_std results in more spread-out and overlapping clusters,
## while a lower value makes the clusters more compact and less overlapping.

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')
cluster_labels = kmeans.fit_predict(X)

# Apply t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X)

# Visualize the original data with cluster labels
plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='viridis', edgecolor='k', s=50)
plt.title("Original Data with Cluster Labels")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# Visualize the data in the t-SNE space
plt.subplot(122)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=cluster_labels, cmap='viridis', edgecolor='k', s=50)
plt.title("t-SNE Visualization of Clusters")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")

plt.tight_layout()
plt.show()
```

### Some key pointers on above coding ( t-SNE )


1. We generate synthetic data with three well-separated clusters using make_blobs from scikit-learn.

2. We apply K-Means clustering to the data to obtain cluster labels.

3. We apply t-SNE to reduce the data to a two-dimensional space (n_components=2) while preserving pairwise similarity relationships. The **perplexity** parameter controls the balance between preserving global and local structures.

4. We visualize the original data with cluster labels in one subplot and the data in the t-SNE space in another subplot.

In the t-SNE visualization, if the clustering quality is good, you should observe well-separated clusters with data points from the same cluster grouped closely together and minimal overlap between clusters. t-SNE helps you check the quality of clustering by providing a visual representation that highlights the separability of clusters.

#### **Quick Note on perplexity parameter in t-Distributed Stochastic Neighbor Embedding (t-SNE)**

The perplexity parameter in t-Distributed Stochastic Neighbor Embedding (t-SNE) is a critical hyperparameter that controls the balance between preserving global and local structures of data points when reducing the dimensionality. Let me explain its role in more detail:

1. **Global Structure:** A lower perplexity value (e.g., 5) emphasizes preserving the global structure of the data. It encourages t-SNE to focus on maintaining the overall layout and relationships between large clusters of data points. As a result, well-separated clusters are preserved, and t-SNE tends to maintain the overall structure of the data. This setting is useful when you want to emphasize the macroscopic view of your data.

2. **Local Structure:** A higher perplexity value (e.g., 50 or more) puts more emphasis on preserving the local structure of the data. It allows t-SNE to pay closer attention to the relationships between nearby data points. This means that t-SNE will work to maintain the fine-grained details and relative positions of data points within clusters. Higher perplexity values are suitable when you want to uncover and visualize fine-grained structures in your data.

In summary, the perplexity parameter in t-SNE allows you to control the trade-off between preserving global and local structures in the reduced-dimensional space. Choosing an appropriate perplexity value depends on the nature of your data and what aspects of the data's structure you want to emphasize in the visualization. It's often a matter of experimentation and domain knowledge to select the best perplexity value for your specific analysis.
